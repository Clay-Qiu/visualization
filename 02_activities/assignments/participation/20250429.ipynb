```python
# generate an random data set
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100,50)
```

```python
# Try to add labels
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('Total Growth Over Time')
ax.set_ylabel('Total Growth')
ax.set_xlabel('Years Since Start')
fig.tight_layout()
```

```python
# Identify font
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}
```

```python
# apply the font
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total Growth Over Time', fontdict=font1)
ax.set_ylabel('Total Growth',fontdict=font2)
ax.set_xlabel('Years Since Start',fontdict=font2)
fig.tight_layout()
```

```python
# change location
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total Growth Over Time', fontdict=font1, loc = 'left')
ax.set_ylabel('Total Growth',fontdict=font2)
ax.set_xlabel('Years Since Start',fontdict=font2)
fig.tight_layout()
```

```python
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
# line plot with both y variables on same axes

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1,y2)
```

```python
# add legends
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right')
fig.show
```

```python
# Moving legend to add shodaw in a block
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right',
frameon = True, #add frame around the legend
fontsize = 12, #change font size
ncol = 2, #specify number of columns
shadow = True )
fig.show()
```

```python
# Moving legend outside of the plot area to avoid covering some data
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left',
bbox_to_anchor =(1, 1))
fig.show()
```

```python
# using scatter plot to show the data and highlight with text function
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.text(10, 95, "This value is important!")
fig.show()
```

```python
# moving the text
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.text(10, 95, "This value is important!",
ha='center', ##alignment of text
color = 'red', ##modify font colour
size = 20) ##modify font size
fig.show()
```

```python
#positioning text annotations with transform function
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])
ax.text(1, 5, ". Data:(1, 5)", transform=ax.transData )
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes )
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure )
```

```python
# add arrow to focus on specific data point
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
arrowprops=dict(facecolor='black', shrink=0.05))
fig.show()
```

```python
#Moving arrow
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
arrowprops=dict(arrowstyle='wedge', color='pink'))
fig.show()
```

```python
#removing tick marks and labels
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
# limit tick marks
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(6))
```

```python
#tick mark intervals
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(8))
```

```python
# rotating axis labels

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xticks(rotation=45, ha='right')
```

```python
# CHange axis font

font1 = {'family':'serif','color':'indigo'}
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xlabel('nex x', fontdict=font1)
```
