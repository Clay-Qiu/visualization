```python
# Load library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# generate an random data set
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100,50)
```

```python
# draw a scatter plot
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y)
```

```python
# draw a bar chart
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
# draw a bar chart vertical
fig, ax = plt.subplots(figsize=(5,3))
ax.barh(x,y)
```

```python
#Line chart
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)
```

```python
#Histogram
fig, ax = plt.subplots(figsize = (5,3))
ax.hist(y)
```

```python
# Try to add labels
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('Total Growth Over Time')
ax.set_ylabel('Total Growth')
ax.set_xlabel('Years Since Start')
fig.tight_layout()
```

```python
# Identify font
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}
```

```python
# apply the font
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total Growth Over Time', fontdict=font1)
ax.set_ylabel('Total Growth',fontdict=font2)
ax.set_xlabel('Years Since Start',fontdict=font2)
fig.tight_layout()
```

```python
# change location
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total Growth Over Time', fontdict=font1, loc = 'left')
ax.set_ylabel('Total Growth',fontdict=font2)
ax.set_xlabel('Years Since Start',fontdict=font2)
fig.tight_layout()
```

```python
# modify data point
fig, ax = plt.subplots(figsize=(5,3))

ax.scatter(x, y,
           marker='*',
           color = "indigo")
fig.show()
```

```python
# modify data point
fig, ax = plt.subplots(figsize=(5,3))

ax.plot(x, y,marker='*',color = "indigo",linestyle = '--',linewidth= 2)
fig.show()
```

```python
# modify data point
fig, ax = plt.subplots(figsize=(5,3))

ax.plot(x, y,marker='*',color = "indigo",linestyle = '--',linewidth= 2, markeredgecolor = '#fa9359', markerfacecolor = '#000000')

ax.grid(axis='y', color = 'blue', linewidth = 2, linestyle = '-.')
```
